Динамические массивы - специфика работы и сложность операций.

Специфика работы с array в Python

Массивы являются основными структурами данных, используемыми в программировании для хранения упорядоченных коллекций элементов одного типа. Они обеспечивают прямой доступ к данным через индексы, что делает их идеальным выбором для обработки последовательностей данных.

Специфика массивов как структур данных

Структура: Массив представляет собой контейнер, заполненный элементами одного типа, расположенными подряд в памяти. Индексация массива начинается с нуля, что означает, что первый элемент имеет индекс 0, второй элемент — индекс 1 и так далее.
Доступ к элементам: Доступ к элементам массива осуществляется напрямую через индекс, что позволяет быстро получить доступ к конкретному элементу.

Изменяемость: Размер массива обычно фиксирован во время его создания, но некоторые языки программирования, например, C++, предоставляют возможность динамического изменения размера массива.

Динамический массив — это массив, размер которого может изменяться во время выполнения программы. В отличие от статических массивов, динамические массивы могут увеличиваться или уменьшаться в зависимости от потребностей программы.

Работа: При добавлении элемента в полный динамический массив, он автоматически перераспределяется на новый блок памяти большего размера, копируя старые элементы и добавляя новый элемент. Этот процесс называется реаллокацией.


Сложность операций:

Добавление элемента: В худшем случае (при необходимости реаллокации) сложность добавления элемента составляет O(n), где n — текущее количество элементов в массиве.

Удаление элемента: Удаление элемента из динамического массива также требует перераспределения элементов, поэтому его сложность также составляет O(n).

Поиск элемента: Поиск элемента по индексу в массиве выполняется за O(1).

Массив array хранит объекты непосредственно, а не ссылки на них. Это настоящий массив, который 
обеспечивает более быстрый доступ и меньший объем памяти. Более быстрый доступ 
обусловлен тем, что нам не нужно проходить через список для обращения к отдельному 
символу. Экономия памяти достигается за счет отсутствия необходимости хранить ссылки 
и объекты отдельно.

Для создания массива мы обращаемся к конструктору array и передаем ему тип элементов 
массива. Размер элементов 
вычисляется автоматически на основе переданных данных(array является динамическим массивом, что означает, что он может изменять свой размер в процессе выполнения программы. Доступ к элементам массива осуществляется по индексу, 
аналогично доступу к элементам списка)

Методы:

append() Этот метод позволяет добавлять новые элементы в конец массива. Особенностью данного метода является то, что добавляемый элемент должен быть того же 
типа, что и уже существующие элементы в массиве. 

count(). Этот метод похож на аналогичный метод, используемый при работе со списками. Он возвращает количество вхождений 
определенного элемента в массиве. Таким образом, можно легко определить, сколько раз 
конкретный элемент встречается в массиве и использовать эту информацию для дальнейшей обработки данных.

insert() предоставляет возможность добавить новый элемент в массив в заданную позицию. Это особенно полезно, когда требуется вставить элемент на определенное 
место в массиве, не нарушая порядок следования других элементов. 



2. a) Используя лямбда-функцию, найдите все числа в заданном списке, которые являются квадратами других чисел.

```python
import random
N = int(input("Введите количество чисел в списке: "))
lst = [random.randint(1, 1000) for _ in range(N)]
print(lst[:15])
```

```python
sqr_in_lst = lambda x: any(x == i**2 for i in range(1, int(x**0.5) + 1)) #лямбда-функция для проверки, является ли число квадратом другого числ
squares = list(filter(sqr_in_lst, lst)) #отбор чисел c помощью filter, которые являются квадратами других чисел

print(squares)
```

2. б) Напишите программу для сортировки заданного списка кортежей по первому элементу каждого кортежа.

```python
tuples_list = [(3, 'a'), (1, 'b'), (2, 'c'), (8, 'i'), (12, 'v'), (0, 'j')] #заданный список кортежей 
sorted_tuples = sorted(tuples_list, key=lambda x: x[0]) #cортировка списка кортежей по первому элементу каждого кортежа

print(sorted_tuples)
```

2. в)Используя лямбда-функцию, найдите все строки в заданном списке строк, которые начинаются и заканчиваются одной и той же буквой.

```python
strings = ['table', 'vav', 'exam', 'surname', 'sure',  'banana'] #заданный список строк

starts_ends_same_letter = lambda s: s[0] == s[-1] #лямбда-функция для проверки, начинается ли строка с первой буквы и заканчивается ли она этой же буквой
same_start_end_strings = list(filter(starts_ends_same_letter, strings)) #ипользование filter для отбора строк, начинающихся и заканчивающихся одной и той же буквой

print(same_start_end_strings)
```

3. a) Используя лямбда-функцию, найдите все числа в заданном списке, которые больше среднего арифметического всех чисел в списке.

```python
N = int(input("Введите количество чисел в списке: "))
numbers = [random.randint(1, 1000) for _ in range(N)]
print(numbers[:15])
```

```python
average = sum(numbers) / len(numbers) #вычисление среднего арифметического всех чисел в списке
is_greater_than_average = lambda x: x > average #лямбда-функция для проверки, больше ли число среднего арифметического
greater_than_average_numbers = list(filter(is_greater_than_average, numbers)) #использование filter для отбора чисел, больших среднего арифметического

print(greater_than_average_numbers)
```

3. б)Напишите программу для сортировки заданного списка строк по длине каждой строки.

```python
words = ['table', 'vav', 'exam', 'surname', 'sure',  'banana'] #заданный список строк
sorted_words_by_length = sorted(words, key=len) #cортировка списка строк по длине каждой строки

print(sorted_words_by_length)
```

3. в)Используя лямбда-функцию, найдите все числа в заданном списке, которые являются кубами других чисел.

```python
N = int(input("Введите количество чисел в списке: "))
lst = [random.randint(1, 10000) for _ in range(N)]
print(lst[:15])
```

```python
sqr_in_lst = lambda x: any(x == i**3 for i in range(1, int(x**(1/3)) + 1)) #лямбда-функция для проверки, является ли число квадратом другого числ
cubes = list(filter(sqr_in_lst, lst)) #отбор чисел c помощью filter, которые являются квадратами других чисел

print(cubes)
```
